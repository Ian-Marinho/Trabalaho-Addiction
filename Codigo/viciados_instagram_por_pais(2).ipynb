# Gráfico de Viciados em Instagram por País
Este notebook mostra apenas os participantes com `Addicted_Score ≥ 8` cujo app mais usado é o **Instagram**.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Carrega os dados
df = pd.read_csv("Filtred_Social_Addiction.csv")

# Filtra viciados com score ≥ 8
viciados = df[df["Addicted_Score"] >= 8]

# Filtra apenas os usuários cujo app mais usado é Instagram
instagram = viciados[viciados["Most_Used_Platform"] == "Instagram"]

# Agrupa por país
uso_insta_por_pais = instagram.groupby("Country").size().reset_index(name="Qtd_Instagram")

# Visualiza
uso_insta_por_pais
```

```python
plt.figure(figsize=(14, 10))
bars = plt.barh(uso_insta_por_pais["Country"], uso_insta_por_pais["Qtd_Instagram"],
                color='lightcoral', edgecolor="black", linewidth=1.5)

for bar in bars:
    width = bar.get_width()
    plt.text(width + 0.5, bar.get_y() + bar.get_height()/2,
             f"{int(width)}", va='center', fontsize=9)

plt.title("Viciados em Instagram por País (score ≥ 8)", fontsize=16)
plt.xlabel("Quantidade de viciados", fontsize=12)
plt.ylabel("País", fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.tight_layout()

# 🔽 Salva e baixa
plt.savefig("grafico_instagram.png", dpi=300, bbox_inches='tight')
from google.colab import files
files.download("grafico_instagram.png")

plt.gca().invert_yaxis()
plt.show()
```
