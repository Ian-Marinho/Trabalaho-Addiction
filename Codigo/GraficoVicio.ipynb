## Graficos

```python
import pandas as pd
import matplotlib.pyplot as plt

# Lê os dados do arquivo CSV (ajuste o nome do arquivo se necessário)
df = pd.read_csv("Filtred_Social_Addiction.csv")

# Filtra viciados com score >= 8
viciados = df[df["Addicted_Score"] >= 8]

# Agrupa por país e plataforma, contando o número de viciados
uso_por_pais = viciados.groupby(["Country", "Most_Used_Platform"]).size().reset_index(name="Qtd_Viciados")

# Ordena para controle visual
dados_agrupados = uso_por_pais.sort_values("Qtd_Viciados", ascending=False)

# Cria o gráfico
plt.figure(figsize=(14, 12))

# Gera barras para cada plataforma separadamente, com controle de espessura
for plataforma in dados_agrupados["Most_Used_Platform"].unique():
    subset = dados_agrupados[dados_agrupados["Most_Used_Platform"] == plataforma]
    plt.barh(
        y=subset["Country"],
        width=subset["Qtd_Viciados"],
        label=plataforma,
        height=0.9  # 👈 aqui você controla a espessura da barra
    )

# Títulos e ajustes
plt.title("Redes sociais mais usadas por jovens viciados (score ≥ 8)", fontsize=16)
plt.xlabel("Quantidade de viciados", fontsize=12)
plt.ylabel("País", fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.legend(title="Plataforma", fontsize=10, title_fontsize=11)
plt.tight_layout()
plt.gca().invert_yaxis()  # Deixa os maiores no topo
plt.show()
```

## Pizza

```python
# Agrupa por país e plataforma, contando quantos viciados usam cada uma
uso_por_pais = viciados.groupby(["Country", "Most_Used_Platform"]).size().reset_index(name="Qtd")
```

```python
import matplotlib.pyplot as plt
import os

# Garante que você esteja usando os dados filtrados corretamente
viciados = df[df["Addicted_Score"] >= 8]
uso_por_pais = viciados.groupby(["Country", "Most_Used_Platform"]).size().reset_index(name="Qtd_Viciados")

# Lista de países a serem analisados
paises = ["India", "USA", "Brazil", "Maldives", "Nepal", "Pakistan", "Russia", "UAE",
          "Mexico", "Italy", "Turkey", "UK", "Poland", "Spain"]

# Cria uma pasta para salvar
os.makedirs("graficos_pizza_iguais", exist_ok=True)

# Função para exibir % e valor real
def make_autopct(values):
    def my_autopct(pct):
        total = sum(values)
        count = int(round(pct * total / 100.0))
        return f'{pct:.1f}%\n({count})'
    return my_autopct

# Gera e salva o gráfico de pizza para cada país com tamanho fixo
for pais in paises:
    dados_pais = uso_por_pais[uso_por_pais["Country"] == pais]

    if dados_pais.empty:
        print(f"Atenção: Nenhum dado encontrado para {pais} com score ≥ 8")
        continue

    plataformas = dados_pais["Most_Used_Platform"]
    quantidades = dados_pais["Qtd_Viciados"]

    plt.figure(figsize=(6, 6))
    plt.pie(quantidades, labels=plataformas,
            autopct=make_autopct(quantidades), startangle=140)
    plt.title(f"Uso de plataformas sociais entre viciados (score ≥ 8) - {pais}")
    plt.axis('equal')
    plt.tight_layout()
    plt.savefig(f"graficos_pizza_iguais/{pais}_pizza.png", dpi=300, bbox_inches='tight', pad_inches=0.2)
    plt.close()
```

## Grafico baixar

```python
import matplotlib.pyplot as plt
import os

# Pasta para armazenar os gráficos
os.makedirs("graficos", exist_ok=True)

# Função para exibir % e valores reais
def make_autopct(values):
    def my_autopct(pct):
        total = sum(values)
        val = int(round(pct * total / 100.0))
        return f'{pct:.1f}%\n({val})'
    return my_autopct

# Lista de países
paises = ["India", "USA", "Brazil", "Maldives", "Nepal", "Pakistan", "Russia", "UAE",
          "Mexico", "Italy", "Turkey", "UK", "Poland", "Spain"]

# Gera gráfico de pizza por país e salva como PNG
for pais in paises:
    dados_pais = uso_por_pais[uso_por_pais["Country"] == pais]
    if dados_pais.empty:
        continue

    plataformas = dados_pais["Most_Used_Platform"]
    quantidades = dados_pais["Qtd_Viciados"]

    plt.figure(figsize=(6, 6))
    plt.pie(quantidades, labels=plataformas,
            autopct=make_autopct(quantidades), startangle=140)
    plt.title(f"Pizza - {pais} (score ≥ 8)")
    plt.axis('equal')
    plt.savefig(f"graficos/{pais}_pizza.png", bbox_inches='tight')
    plt.close()
```

```python
# Organiza os dados para o gráfico de linhas (barras horizontais)
dados_agrupados = uso_por_pais.sort_values("Qtd_Viciados", ascending=False)

plt.figure(figsize=(14, 12))

for plataforma in dados_agrupados["Most_Used_Platform"].unique():
    subset = dados_agrupados[dados_agrupados["Most_Used_Platform"] == plataforma]
    plt.barh(
        y=subset["Country"],
        width=subset["Qtd_Viciados"],
        label=plataforma,
        height=0.9
    )

plt.title("Gráfico de barras - plataformas mais usadas por viciados (score ≥ 8)", fontsize=16)
plt.xlabel("Quantidade de viciados", fontsize=12)
plt.ylabel("País", fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.legend(title="Plataforma", fontsize=10, title_fontsize=11)
plt.tight_layout()
plt.gca().invert_yaxis()
plt.savefig("graficos/barras_horizontal.png", bbox_inches='tight')
plt.close()
```
